This notebook will read in any JSON file and parse the data as Twitter tweet objects. Please see Twitter reference for more: [Tweet object](https://developer.twitter.com/en/docs/tweets/data-dictionary/overview/tweet-object).

Let's import all the required libraries:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
from textblob import TextBlob
import html
from datetime import datetime
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
from collections import Counter
import string
````

This is where we are creating Pandas **DataFrame** object and assigning it to a Python dictionary (called **Series** in Pandas). The lines attribute tells the read_json function to read in the JSON file as a series of lines.

To make things easier for us, we are assigning different attributes of the Tweet object to different variables.
Please feel free to input different json files (Twitter format).

````python
filename = 'fifa.json'
tweets = pd.read_json(filename,orient='columns',lines='True')

content_cols = ['id','text','created_at','source','favorite_count','retweet_count','lang']
content = pd.DataFrame(data=tweets,columns=content_cols)

json_string = tweets['user'].to_json()
user_data = pd.read_json(json_string,orient='index')

user_cols = ['id','name','screen_name','location','followers_count','friends_count','favourites_count','statuses_count']
user = pd.DataFrame(data=user_data,columns=user_cols)
````

### Relationship between Followers and Friends
Twitter defines **followers** as users following the specific user, whereas **friends** are every user the specific user is following. The line gets blurry with the mechanical nature of following/followers in Twitter. However, we can create a Simple Linear Regression Model (SLR) to observe any relationship between followers and friends. This can help us to decide if a particular follower maybe a bot or not.

We are choosing the number of followers as the _independent variable (IV)_ and the number of friends as the _dependent variable (DV)_.

````python
sns.set()
fig_dims = (11.7,8.27)
fig,ax = plt.subplots(figsize=fig_dims)
sns.regplot(y="followers_count",x="friends_count",data=user,ax=ax)
plt.title("Followers vs. Friends")
plt.xlabel("Followers")
plt.ylabel("Friends")
plt.ylim(0,None)
plt.xlim(0,None)
sns.despine()
````

#### Distribution of Tweet length
We are going to plot a univariate kernel density estimate (KDE). The goal is to estimate the probability density function of tweets.

````python
tweet_len = []

for k in content['text']:
    if isinstance(k,str):
        tweet_len.append(len(k))

sns.set()
fig_dims = (11.7,8.27)
fig,ax = plt.subplots(figsize=fig_dims)
sns.kdeplot(tweet_len,shade=True)
plt.title("Distribution of Tweet length")
plt.xlabel("No. of characters")
plt.ylim(0,None)
plt.xlim(0,None)
sns.despine()
````

In the plot above, we can see that most of the tweets reside around the 140 characters mark. It is due to Twitter's limitation of  using only 140 characters.

### Sentiment Analysis Using TextBlob
Sentiment analysis will allow us to gauge the positivity or negativity of a tweet. To achieve this, I have downloaded the default NLTK corpa and using the polarity function of the sentiment object. If the polarity is more than zero (0), then we label it as being **_positive_**. If the polarity equals zero, we label it as being **_neutral_**, and last but not the least, we assigned a **_negative_** label to a polarity of less than zero.

#### Pre-processing the Tweet and Parsing it
We first use the str function to convert the tweet to a _string_ object. Then, we use the html parsing library that comes default with Python to un-escape the tweets. When we retrieved the Tweets, it was already escaped so it contained characters such newline (\n), tab (\t), etc. Unescaping let us revert those characters back to the correct encoding.

**Using regex for parsing**

The following line uses the re library of Python and gets rid of @mentions, retweets, hashtags, links, emoticons:
```
tweet_content[key] = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|('https?://[A-Za-z0-9./]+',')", " ", tweet_content[key]).split())

```

````python
def get_time(tweet):
    return datetime.strptime(str(tweet), "%Y-%m-%d %H:%M:%S")

def get_sentiment(tweet):
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1

def get_subjectivity(tweet):
    analysis = TextBlob(tweet)
    return analysis.sentiment.subjectivity


def sanitize(tweet):
        tweet = str(tweet)
        tweet = html.unescape(tweet)
        tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|('https?://[A-Za-z0-9./]+',')", " ", tweet).split())
        return tweet
    
parsed_text = pd.DataFrame(data=[sanitize(s) for s in content['text']],columns=['Tweet'])
parsed_text['Sentiment'] = np.array([get_sentiment(k) for k in parsed_text['Tweet']])

pos_tweets = [tweet for index,tweet in enumerate(parsed_text['Tweet']) if parsed_text['Sentiment'][index] > 0]
neu_tweets = [tweet for index,tweet in enumerate(parsed_text['Tweet']) if parsed_text['Sentiment'][index] == 0]
neg_tweets = [tweet for index,tweet in enumerate(parsed_text['Tweet']) if parsed_text['Sentiment'][index] < 0]

# print("Percentage of positive tweets: {}%".format(len(pos_tweets)*100/len(parsed_text['Tweet'])))
# print("Percentage of neutral tweets: {}%".format(len(neu_tweets)*100/len(parsed_text['Tweet'])))
# print("Percentage of negative tweets: {}%".format(len(neg_tweets)*100/len(parsed_text['Tweet'])))

pie_labels = ['Positive','Neutral','Negative']
pie_sizes = [len(pos_tweets),len(neu_tweets),len(neg_tweets)]
colors = ['#2ecc71','#3498db','#e74c3c']
fig_dims = (11.7,8.27)
fig,ax = plt.subplots(figsize=fig_dims)
plt.pie(pie_sizes,labels=pie_labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title("Sentiment Percentage")
plt.show()
````

### Tokenization & Frequency Terms

The tweets that we have collected has lots of materials that we are not interested in. Some of them include English stopwords (such as, "the","from","a", etc.), punctuations, whitespaces. To deal with the problem, we have used the tokenizer library from NLTK to separate the words into their individual spaces. Then, we have matched the case of the words by lowering all of them. 

We then used a Counter object to count the number of times a particular term appears in the Twitter data. After that, we created a Pandas DataFrame and utilized Seaborn to plot a bar chart. 

````python
tokens = []
for key,value in parsed_text['Tweet'].iteritems():
    tokens.append(word_tokenize(value))

lower_case = []
for k in tokens:
    for j in k:
        lower_case.append(j.lower())
        
table = str.maketrans('','',string.punctuation)
stripped = [w.translate(table) for w in lower_case]
words = [word for word in stripped if word.isalpha()]

stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
word_counts = Counter(words)

print(words)
lst = word_counts.most_common(10)
df = pd.DataFrame(lst,columns=['Word','Count'])

sns.set()
fig_dims = (11.7,8.27)
fig,ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='Word',y='Count',data=df)
plt.title("Top 10 Most Used Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.ylim(0,None)
plt.xlim(0,None)
sns.despine()
````
